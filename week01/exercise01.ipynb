```python
import matplotlib.pyplot as plt
import math
import random

def generate_random_normal():
    u1, u2 = random.random(), random.random()
    z0 = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
    return z0

def normal_distribution_density(x, mu=0, sigma=1):
    return (1 / (math.sqrt(2 * math.pi) * sigma)) * math.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

sample_size = 10000
samples = [generate_random_normal() for _ in range(sample_size)]

plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

xmin, xmax = plt.xlim()
x = [xmin + (xmax - xmin) * i / 100 for i in range(101)]
p = [normal_distribution_density(xi) for xi in x]
plt.plot(x, p, 'k', linewidth=2)

plt.title("Histogram Fit to Normal Distribution")
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```
